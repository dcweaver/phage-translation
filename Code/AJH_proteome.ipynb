```python
from Bio import Seq,SeqIO
import pandas as pd
import glob
```

```python
for virus_tsv_file in glob.glob('../Data/287_rep_viruses/*.clean.tsv'):
    virus_df = pd.read_csv(virus_tsv_file, sep='\t', index_col=0)
    virus_name = virus_tsv_file.split('/')[-1].split('.')[0]
    print('###', virus_tsv_file)
    print(virus_name, virus_df.shape)
    
    with open('/Users/adamhockenberry/workspace/PHACTS/{}.faa'.format(virus_name), 'w') as outfile:
        for index in virus_df.index:
            identifier = virus_df.at[index, 'viral_id']
            cds = virus_df.at[index, 'coding_sequence']
            seq = Seq.Seq(cds)
            outfile.write('>{}\n'.format(identifier))
            outfile.write('{}\n'.format(str(seq.translate())))
```
