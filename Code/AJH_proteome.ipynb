```python
from Bio import Seq,SeqIO
import pandas as pd
```

```python
virus_df = pd.read_csv('../Data/562_rep_viruses/6088.clean.tsv', sep='\t', index_col=0)
print(virus_df.shape)
virus_df.head()
```

```python
with open('/Users/adamhockenberry/Desktop/temp_virus.faa', 'w') as outfile:
    for index in virus_df.index:
        identifier = virus_df.at[index, 'viral_id']
        cds = virus_df.at[index, 'coding_sequence']
        seq = Seq.Seq(cds)
        outfile.write('>{}\n'.format(identifier))
        outfile.write('{}\n'.format(str(seq.translate())))
```
